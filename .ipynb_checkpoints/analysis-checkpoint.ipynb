```python
import pandas as pd
```

```python
esl = pd.read_csv('eslKillFeed.csv')
esl
```

```python
placement_points = {1:65,2:58,3:56,4:54,5:52,6:50,7:48,8:46,9:44,10:42,11:40,12:39,13:38,14:37,15:36,16:35,17:34,18:33,19:32,20:31,21:30,22:29,23:28,24:27,25:26,26:25,27:24,28:23,29:22,30:21,31:20,32:19,33:18,34:17,35:16,36:15,37:14,38:13,39:12,40:11,41:10,42:9,43:8,44:7,45:6,46:5,47:4,48:3,49:2,50:1}
```

```python
placements = pd.DataFrame(index = esl['Player'].unique(), columns=['Points', 'Victory Royales', 'Kill Points', 'Kills', 'Placement Points']).fillna(0)
```

```python
for i, row in esl.iterrows():
    placements['Kills'][row['Player']] += row['Kills']
    placements['Points'][row['Player']] += row['Kills']*5
    placements['Kill Points'][row['Player']] += row['Kills']*5
    if (row['Placed'] < 51):
        placements['Points'][row['Player']] += placement_points[row['Placed']]
        placements['Placement Points'][row['Player']] += placement_points[row['Placed']]
    if (row['Placed']==1):
        placements['Victory Royales'][row['Player']] +=1
placements['Avg. Placement'] = esl.groupby('Player')['Placed'].mean()
placements = placements.sort_values(['Points','Victory Royales','Kills', 'Avg. Placement'], ascending=[False, False, False, True]).reset_index()
placements.index+=1
```
